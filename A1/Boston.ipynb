```python
data = load_boston(return_X_y=False)

print("Typ von data",type(data))
print()
print("Offizielle Beschreibung des Inhalts:\n\n",data.DESCR)
```

Pandas Dataframe erstellen

```python
boston_housing = pd.DataFrame(data['data'], columns=data['feature_names'])
# zeige die ersten 5 Datensätze
boston_housing.head()
```

```python
boston_housing = pd.DataFrame(data['data'])
boston_housing.columns=data['feature_names']
# zeige die letzten 10 Datensätze
boston_housing.tail(10)
```

```python
# ergänze den Preis (y-Wert)
boston_housing['PRICE'] = data['target']
boston_housing.head(10)
```

Zimmeranzahl vs Preis

```python
plt.figure(figsize=[15,8])
ax = sns.scatterplot(x="RM", y= 'PRICE',data=boston_housing)
ax.set_title('Zimmeranzahl vs Preis')
ax.set_xlabel('Zimmeranzahl')
ax.set_ylabel('Preis')
plt.show()
```

```python
pd.crosstab(boston_housing['RM'], boston_housing['PRICE'])
```

```python
boston_housing['RM_BIN'] = round(boston_housing['RM'],0)
boston_housing['PRICE_BIN'] = pd.cut(boston_housing['PRICE'],
                                     bins=4,
                                     labels=["günstig", "etwas teurer", "teuer", "richtig teuer"])
boston_housing.head()
```

```python
boston_housing.info()
```

```python
boston_housing.describe()
```

```python
boston_housing.describe(include='category') # include='all'
```

```python
pd.crosstab(boston_housing['RM_BIN'], boston_housing['PRICE_BIN'], margins='all')
```

```python
plt.figure(figsize=[8,6])
ax = sns.heatmap(pd.crosstab(boston_housing['RM_BIN'], boston_housing['PRICE_BIN']),
            cmap="YlGnBu",
           linewidths=.5,
           annot=True,
           fmt=".0f")
ax.set_title('Anzahl vs Preis')
ax.set_xlabel('Zimmeranzahl')
ax.set_ylabel('Preis')
plt.show()
```

```python
plt.figure(figsize=[12,8])
ax = sns.countplot(x="RM_BIN",
              data=boston_housing,
              hue="PRICE_BIN",
             palette="PiYG_r")
ax.set_title('Anzahl vs Preis')
ax.legend(title='Preiskategorie')
ax.set_xlabel('Anzahl Zimmer')
ax.set_ylabel('Anzahl Bezirke')
plt.show()
```

Zimmeranzahl + Stickoxidbelastung vs Preis

```python
boston_housing[['RM', 'NOX', 'PRICE']].describe()
```

```python
plt.figure(figsize=[15,8])
ax = sns.scatterplot(x="RM", y= 'PRICE',data=boston_housing,
                hue="NOX", size="NOX",
                palette='PiYG_r')
ax.set_title('Zimmeranzahl und Stickoxidbelastung vs Preis')
ax.legend(title='Stickoxodbelastung')
ax.set_xlabel('Anzahl Zimmer')
ax.set_ylabel('Preis')
plt.show()
```
